```python
# ファイルのエンコーディングを推測
import chardet

with open('Speed Dating Data.csv', 'rb') as f:
    result = chardet.detect(f.read())
    print(f"推定エンコーディング: {result['encoding']}")
    print(f"信頼度: {result['confidence']}")
```
